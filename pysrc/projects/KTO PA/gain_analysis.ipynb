```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
# Define parameters
C = 10e-9       # Capacitance 10 nF
Z = 50          # Impedance 50 Ohm
L_0 = 10e-9     # Base inductance 10 nH
omega = 100e6   # Frequency for the sinusoidal functions 100 MHz
```

```python
def calc_thoretical_gain(L_p, C=C, Z=Z, L_0=L_0, omega=omega):
    K = 1/(C*Z)
    gain = (((2*K/omega) + L_p/L_0)/((2*K/omega) - L_p/L_0))**2
    return gain

def plot_gain(start, end, step):
    L_p = np.linspace(start, end, step)
    gain_array = [calc_thoretical_gain(L_p=l) for l in L_p]
    plt.plot(L_p, gain_array)
    plt.xlabel('L_p [H]')
    plt.ylabel('Gain')
    plt.title(f'Theoretical gain vs L_p ({start} to {end} H)')
    plt.show()
```

```python
plot_gain(3.9e-10, 4.1e-10, 10000)
plot_gain(3.9e-10, 3.99e-10, 10000)
```
